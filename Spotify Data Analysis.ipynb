# Setting release date as new index

```python
df.set_index("release_date", inplace=True)
df.index=pd.to_datetime(df.index)
df.head()
```

# convert the values in duration columns from milisecond to seconds

```python
df['duration']=df['duration_ms'].apply(lambda x: round(x/1000))
df.drop('duration_ms',inplace=True,axis=1)
```

```python
df['duration'].head()
```

# Correlation Heatmap Between Variables

```python
corr_df = df.drop(['key', 'mode', 'explicit'], axis=1).corr(method='pearson')
plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt=".1g", vmin=-1, vmax=1, center=0, cmap="inferno", linewidths=1, linecolor="black")
heatmap.set_title("Correlation Heatmap Between Variables")
plt.show()
```

here there is strong +ve corr between energy and loudness
and strong -ve corr between energy and acoustiness

as we have a large dataset getting a graph is getting too much time so we shorten the dataset in below code

```python
sample_df=df.sample(int(0.004*len(df)))
len(sample_df)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y='loudness',x='energy',color='c').set(title="loudness vs Energy Corr")
```

the more the loudness  the  more the energy (insights) (+VE regression)

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df, y='popularity',x='acousticness',color='b').set(title="popularity vs acoustisness")
```

```python
there is -Ve regression as popularity increases acousticness decreases
```
